```python
value = int(input('정수를 입력하세요 :'))
print(value)
```

```python
if value % 2 == 0:
  print("짝수")
elif value % 2 == 1:
  print("홀수")
```

```python
match value % 2:
  case 0:
    print("짝수")
  case 1:
    print("홀수")
  case _ :  # Java의 default
    print("기타")
```

```python
age = 18
status = '성인' if age >= 20 else '미성년자'
print(status)
```

```python
numbers = [1, 2, 3, 4, 5]
new_list = ['짝수' if ele % 2 == 0 else '홀수' for ele in numbers]
print(new_list)
```
