```python
# position parameter
def add(a=0, b=0):
  return a + b
```

```python
result = add(10, 20)
print(result)
```

```python
result = add(30)
print(result)
```

```python
result = add(b=100, a=200)
print(result)
```

```python
# keywords parameter
result = add(b=100, a=200)
print(result)
```

```python
# Variable Positional Parameters
# 매개 변수는 함수 내에서 튜플 객체로 사용
def sum_numbers(*args):
  print(type(args))
  print(args)
  sum = 0
  for ele in args:
    sum+= ele
  return sum
```

```python
print(f'총합 : {sum((1, 2, 3, 4))}')
```

```python
# Variable Keyword Parameters
# 매개 변수는 함수 내에서 딕셔너리 객체로 사용
def print_keyword_arguments(**kwargs):
    person = {
        'name' : '이름',
        'age' : 10,
        'address' : '서울'
    }

    print(kwargs, type(kwargs))

    # 매개 변수 출력
    for key, value in kwargs.items():
        print(f'{key} : {value}')

    # person 객체 출력
    for key, value in person.items():
        print(f'{key} : {value}')
```

```python
print_keyword_arguments(name='홍길동', age=30, address='부산', gender='남')
```
