```python
# dictionary : 키, 값의 쌍으로된 요소를 저장하고자 할 때 상요하는 자료구조 / 키는 중복 불가, 순서X
person = {'name':'홍길동', 'age':'20', 'addres':'부산'}
person
```

```python
print(type(person))
```

```python
dir(person)
```

```python
# 특정 키의 값 접근
print(person['name'])
print(person.get('name'))
```

```python
# 요소 추가
person['tel'] = '010-1234-5678'
person
```

```python
# 요소 삭제
del person['tel']
person
```

```python
x, y = (10, 20)
print(x, y)
x, y = [10, 20]
print(x, y)
x, y = {10, 20}
print(x, y)
x, y = {'1':10, '2':20} # key 값이 개별 변수에 저장
print(x, y)
```

```python
# 요소 순회
person.items()  # 키, 값을 튜플 요소로 갖는 리스트로 반환

for key, value in person.items():
  print(key, value)
```

```python
person.keys() # 딕셔너리에서 키만 리스트로 반환
for key in person.keys():
  print(key)
```

```python
person.values() # 딕셔너리에서 값만 리스트로 반환
for value in person.values():
  print(value)
```
